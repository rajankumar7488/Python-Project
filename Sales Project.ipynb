```python
csv.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# drop null values
csv.dropna(inplace=True)
```

```python
# change data type
csv['Amount'] = csv['Amount'].astype('int')
```

```python
# use describe() for specific columns
csv[['Age', 'Orders', 'Amount']].describe()
```

```python
# plotting a bar chart for Gender and it's count

var = sns.countplot(x = 'Gender',data = csv)

for bars in var.containers:
    var.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = csv.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
var = sns.countplot(data = csv, x = 'Age Group', hue = 'Gender')

for bars in var.containers:
    var.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = csv.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
# total number of orders from top 10 states

sales_state = csv.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = csv.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
var = sns.countplot(data = csv, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in var.containers:
    var.bar_label(bars)
```
